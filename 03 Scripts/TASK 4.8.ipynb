#### 4 Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# creating new column based on the max amount of orders
df_orders_products_merged_new['max_orders'] = df_orders_products_merged_new.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_orders_products_merged_new.head()
```

```python

# 1 defining flags for the loyality check
def customers_loyalty(row):

    if row['max_orders'] > 40:
    return 'Loyal Customer'
  elif (row['max_orders'] > 10) and (row['max_orders'] <= 40):
    return 'Regular Customer'
  elif row['max_orders'] <= 10:
    return 'New Customer'
  else: return 'Not enough data'
```

```python
# creating new column for loyaility check based on the flags deifined earlier. New columns are created using loc method
df_orders_products_merged_new.loc[df_orders_products_merged_new['max_orders'] > 40, 'customers_loyalty'] = 'Loyal customer'
```

```python
df_orders_products_merged_new.loc[(df_orders_products_merged_new['max_orders'] <= 40) & (df_orders_products_merged_new['max_orders'] > 10), 'customers_loyalty'] = 'Regular customer'
```

```python
df_orders_products_merged_new.loc[df_orders_products_merged_new['max_orders'] <= 10, 'customers_loyalty'] = 'New customer'
```

```python
# checking if the new column created and works correctly
df_orders_products_merged_new.head().append(df_orders_products_merged_new.tail())
```

#### 6 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

#####  If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

##### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating new column which will show average price
df_orders_products_merged_new['mean_price'] = df_orders_products_merged_new.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# checking if the column appearing in the data frame 
df_orders_products_merged_new.head()
```

```python
# creating a flag that can be used for creating of a new column
# 1 defining flags for the loyality check
def spending_characteristic(row):

  if row['mean_price'] < 10:
    return 'Low spender'
  elif row['mean_price'] >= 10:
    return 'High spender'
  else: return 'Not enough data'
```

```python
# creating additional column using col method
df_orders_products_merged_new.loc[df_orders_products_merged_new['mean_price'] < 10, 'spending_characteristic'] = 'Low spender'
```

```python
df_orders_products_merged_new.loc[df_orders_products_merged_new['mean_price'] >= 10, 'spending_characteristic'] = 'High spender'
```

```python
# checking if the new column was added
df_orders_products_merged_new.head()
```

#### 7  In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

#### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# creating another column with median of the days since prior order column
df_orders_products_merged_new['since_priors_order_median'] = df_orders_products_merged_new.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# checking if the new column was created 
df_orders_products_merged_new.head()
```

```python
# creating flag for a new column
def ordering_frequency(row):

    if row['since_priors_order_median'] > 20:
    return 'Non-frequent customer'
  elif (row['since_priors_order_median'] >10) and (row['since_priors_order_median'] <= 20):
    return 'Regular customer'
  elif row['since_priors_order_median'] <= 10:
    return 'Frequent customer'
  else: return 'Not enough data'
```

```python
# creating new columns using loc method
df_orders_products_merged_new.loc[df_orders_products_merged_new['since_priors_order_median'] > 20, 'ordering_frequency'] = 'Non-frequent customer'
```

```python
df_orders_products_merged_new.loc[(df_orders_products_merged_new['since_priors_order_median'] >10 )  & (df_orders_products_merged_new['since_priors_order_median'] <= 20), 'ordering_frequency'] = 'Regular customer'
```

```python
df_orders_products_merged_new.loc[df_orders_products_merged_new['since_priors_order_median'] <= 10, 'ordering_frequency'] = 'Frequent customer'
```
