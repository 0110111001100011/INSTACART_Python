```python
# creating loyality flags
# 1 defining flags for the loyality check
def customers_loyalty(row):

  if row['max_order'] > 40:
    return 'Loyal Customer'
  elif (row['max_order'] > 10) and (row['max_order'] <= 40):
    return 'Regular Customer'
  elif row['max_order'] <= 10:
    return 'New Customer'
  else: return 'Not enough data'
```

```python
# applying the same to the sub data frame
df['max_order']= df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#adding new column to the subset of the df
df['customer_loyalty'] = df.apply(customers_loyalty, axis=1)
```
