```python
# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. 
# Now, they want “Busiest day” to become “Busiest days” (plural).
# This label should correspond with the two busiest days of the week as opposed to the single busiest day.
# At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.
```

```python
# creating shell that we can use as a new column results.
result_busiest_days = []

for value in df_orders_products_merged['order_dow']:
  if value in [0,1]:
    result_busiest_days.append('Busiest days')
  elif value in [3,4]:
    result_busiest_days.append('Slowest days')
  else:
    result_busiest_days.append('Regularly busy')
```

```python
# creating new column with the busiest days and slowest days
df_orders_products_merged['busiest_days_new'] = result_busiest_days
```

```python
# checking the results
df_orders_products_merged['busiest_days_new'].value_counts(dropna = False)
```

```python
#When too many users make Instacart orders at the same time, the app freezes.
#The senior technical officer at Instacart wants you to identify the busiest hours of the day.
#Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.”
#Create a new column containing these labels called “busiest_period_of_day.”

df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# creating a shell for the busiest hour column
result_busiest_hour = []
for value in df_orders_products_merged['order_hour_of_day']:
  if value in [10,11,14,15,13,12,16,9]:
    result_busiest_hour.append('Most orders ')
  elif value in [17,8,18,19,20,7,21,22]:
    result_busiest_hour.append('average orders')
  elif value in [23,6,0,1,5,2,4,22]:
    result_busiest_hour.append('fewest orders')
  else:
    result_busiest_hour.append('Regularly busy')
```

```python
# adding new column to the data frame
df_orders_products_merged['busiest_hours'] = result_busiest_hour
```

```python
# checking frequency of the new column
df_orders_products_merged['busiest_hours'].value_counts()
```
